```python
import skimage.io
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("processed_data/small_generic/df.csv")
df["id"].head()
```

```python
img_folder_path = "processed_data/image_pool/{0}_{0}/".format(256)
img_paths = list(map(lambda id: img_folder_path + str(id) + ".png", df["id"].values))
img_paths[1]
```

```python
images = skimage.io.imread_collection(img_paths[:3])
list(images)
```

```python
plt.imshow(images[0])
```

```python
vq_dim = 16
quantized_chuncks = []
img_chuncks = np.split(images[0], vq_dim, axis=1)

for chunck in img_chuncks:
    tiles = np.split(chunck, vq_dim, axis=0)
    print(tiles[0].shape)
    quantized_chuncks.extend(tiles)
```
